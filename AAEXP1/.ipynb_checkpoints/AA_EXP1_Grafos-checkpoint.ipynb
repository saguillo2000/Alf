  <p>
 Como se ha visto en clase, podemos emplear grafos dirigidos para representar flujos. En particular, hemos visto el algoritmo de Ford-Fulkerson que encuentra el flujo máximo de una red dada. Dado el pseudo-código que se incluía en las trasparencias de teoría (también aquí), se os pide que implementéis dicho algoritmo.

````python
def path(G, origen, destino):
    
    #Solucion de Adria Saiz 
    
    path = []

    
    visited = {}
    previous = {}
    
    for node in G.nodes():
        visited[node] = False
        previous[node] = None
    
    stack = [origen]
    trobat = False
    
    
    
    while ((len(stack) != 0) and (not trobat)):
        
        nodoActual = stack.pop()
        
        vecinos = [nodoVecino 
                   for nodoVecino in G.neighbors(nodoActual) 
                   if not visited[nodoVecino]]
        
        for nodoVecino in vecinos:
            
            visited[nodoVecino] = True
            stack.append (nodoVecino)
            previous[nodoVecino] = nodoActual
            
            if (nodoVecino == destino):
                trobat = True
                
        visited[nodoActual] = True
    
    
    path = [destino]
            
    while previous[path[0]] != origen:
        path = [previous[path[0]]] + path
            
    path = [origen] + path

    return path 
````

````python
'''
Dado un grafo y el flujo actual que circula por el mismo, esta función 
establece si algún camino todavía es posible.
'''
def exists_path(G, origen, destino):
    
    path = []

    
    visited = {}
    previous = {}
    
    for node in G.nodes():
        visited[node] = False
        previous[node] = None
    
    stack = [origen]
    trobat = False
    
    
    
    while ((len(stack) != 0) and (not trobat)):
        
        nodoActual = stack.pop()
        
        vecinos = [nodoVecino 
                   for nodoVecino in G.neighbors(nodoActual) 
                   if not visited[nodoVecino]]
        
        for nodoVecino in vecinos:
            
            visited[nodoVecino] = True
            stack.append (nodoVecino)
            previous[nodoVecino] = nodoActual
            
            if (nodoVecino == destino):
                trobat = True
                
        visited[nodoActual] = True

    return trobat    

'''
Dado un grafo dirigido, cuyas aristas tienen un atributo 'capacity', 
el cual indica cuánto flujo puede circular por ellas, un nodo fuente (source) y un nodo sumidero (target),
este algoritmo calcula el flujo máximo que puede circular desde s hasta t.
'''
def ford_fulkerson(G,s,t):
    max_flow = 0
    # Implementad aquí vuestra solución
    
    path_aux = []
    
    
    for edge in G.edges():
        G.edges[edge]['flux'] = 0

        
        
        
    while exists_path(G, s, t):
        
        #Creamos un camino que va de s a t
        path_aux = path(G, s, t) 
        
        min_capacity = 9999
        
        #A continuación averiguaremos cual es el valor de cf(path)
        #compararemos todas las capacity del camino
        
        for i in range(len(path_aux)-1):
            if(G.edges[(path_aux[i],path_aux[i+1])]['capacity']<min_capacity):
                min_capacity = G.edges[(path_aux[i],path_aux[i+1])]['capacity']

        
        
        #Aquí haremos lo siguiente:
        #1. Crearemos una arista inversa a todas la aristas de path(Si esta no existe) y le sumaremos el cf
        #2. Si por alguna razón la capacity de alguna arista es 0 o menos(no puede ser menos pero por si acaso)
        #   se elimina dicha arista
        
        #De esta forma tendremos un grafo inconexo con una parte, ahora debemos ver cual es el flujo minimo de t.
        for i in range(len(path_aux)-1):
            G.edges[(path_aux[i],path_aux[i+1])]['flux'] += min_capacity
            if((path_aux[i+1],path_aux[i]) not in G.edges):
                G.add_edge(path_aux[i+1],path_aux[i])
                G.edges[(path_aux[i+1],path_aux[i])]['capacity'] = min_capacity
            else:
                G.edges[(path_aux[i+1],path_aux[i])]['capacity'] += min_capacity



            G.edges[(path_aux[i],path_aux[i+1])]['capacity'] -= min_capacity
            if (G.edges[(path_aux[i],path_aux[i+1])]['capacity'] <= 0):
                G.remove_edge(path_aux[i],path_aux[i+1])
 
    #Averigua cual es el Max_Flow
    for n in G.neighbors(t):
         max_flow += G.edges[(t,n)]['capacity']
    
    
    return max_flow
````
